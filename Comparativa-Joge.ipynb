##2. Establecemos una comparativa entre todos los modelos creados

```python
import os
path_modelos_yangu = path + '1Yanguas/'
path_modelos_golvi = path + '1Golvin/'
path_modelos_jorge = path + '1Jorge/'
paths = [path_modelos_jorge]


dfs_confres = []

for path_usuario in paths:
  lista_dir = os.listdir(path_usuario)
  for modelo in lista_dir:
    try:
      df_result = pd.read_excel(path_usuario + modelo + '/informe.xlsx')
      df_config = pd.read_excel(path_usuario + modelo + '/model_conf.xlsx')
      df_config['precision_test'] = df_result.iloc[4, 1]
      df_config['Modelo'] = modelo
      dfs_confres.append(df_config)
    except: pass

df_resultados = pd.concat(dfs_confres, )
df_resultados.set_index('Modelo', inplace=True) 
df_resultados.drop(df_resultados.columns[0], axis=1, inplace=True)
df_resultados.sort_values(by=['precision_test'], inplace=True, ascending=False)
df_resultados
```
